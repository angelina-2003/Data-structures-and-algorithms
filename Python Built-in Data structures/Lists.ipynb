### Remove elements 

`Remove()` will remove the first occurrence of the element.

```python
fruits = ["apple", "banana", "cherry", "mango", "banana"]

fruits.remove("banana")

print(fruits)
```

`pop()` method removes the specific index. If index not specified, it will remove the last item.

```python
fruits.pop(1)   # this will remove cherry
    
print(fruits)
```

```python
fruits.pop()        # this alone returns banana
print(fruits)       # this prints list after removing last element, which is banana
```

## Loop Lists

### 1. Looping through a list:



```python
fruits = ["apple", "banana", "cherry", "mango", "banana"]

for x in fruits:
    print(x)
```

### 2. Looping through index numbers

```python
for i in range(len(fruits)):
    print(fruits[i])
```

### 3. While loop

```python
i = 0

while i < len(fruits):
    print(fruits[i])
    i += 1
```

### 4. Looping using List Comprehension

List Comprehension offers the shortest syntax for looping through lists:

```python
[print(x) for x in fruits]
```

## Sorting Lists

### 1. Sort List Alphanumerically 

List objects have a `sort()` method that will sort the list alphanumerically, ascending, by default:


```python
people = ["John", "Ben", "Doe", "Smith", "Angie"]

people.sort()

print(people)
```

### 2. Sorting list Numerically



```python
nums = [100, 50, 65, 82, 23]
nums.sort()
print(nums)
```

### 3. Sort Descending 


```python
fruits = ["orange", "mango", "kiwi", "pineapple", "banana"]
fruits.sort(reverse = True)
print(fruits)
```

### 4. Case insensitive sort

By default the sort() method is case sensitive, resulting in all capital letters being sorted before lower case letters:

Luckily we can use built-in functions as key functions when sorting a list.

So if you want a case-insensitive sort function, use str.lower as a key function:


```python
fruits = ["banana", "Orange", "Kiwi", "cherry"]
fruits.sort(key = str.lower)
print(fruits)
```

### 5. Reverse Order

What if you want to reverse the order of a list, regardless of the alphabet?

The `reverse()` method reverses the current sorting order of the elements.

```python
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.reverse()
print(thislist)
```

## Joining Lists

### 1. Using +


```python
list1 = ["a", "b", "c"]
list2 = [1, 2, 3]

list3 = list1 + list2
print(list3)
```

### 2. Using append

But append does it one by one so we need a for loop

```python
list1 = ["a", "b" , "c"]
list2 = [1, 2, 3]

for x in list2:
  list1.append(x)

print(list1)
```

### 3. Using Extend() Method

Simply use the extend() method to add the elements of list2 to list1:

```python
list1 = ["a", "b" , "c"]
list2 = [1, 2, 3]

list1.extend(list2)
print(list1)
```
