## The problem recursion solves:

Some data is nested, it goes as deep as it wants and you don't know how deep in advance.

You can't write a fixed loop for this because you don't know how many levels deep it goes.

Here's the data structure:

```python
epic = {
    "title": "Build Auth System",
    "children": [
        {
            "title": "Login page",
            "children": [
                {"title": "Build form", "children": []},
                {"title": "Validate input", "children": []}
            ]
        },
        {
            "title": "Logout",
            "children": [
                {"title": "Clear session", "children": []}
            ]
        }
    ]
}
```

## Task:

Write a function `print_all_tasks(task, level=0)` that prints every task title, indented by its level:


**Hint** : the function should call itself when it finds children.

```python
def print_all_tasks(task, level=0):
    print(" " * level + task["title"])

    for child in task["children"]:
        print_all_tasks(child, level+1)


print_all_tasks(epic)
```
