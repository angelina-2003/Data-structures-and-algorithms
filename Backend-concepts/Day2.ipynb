## The scenario:

You are building a tiny piece of a project management tool (think jira lite).

You have users. Each user has a name, email and a list of assigned tasks.

You need to:
- Store and look up users quickly
- Add tasks to a specific user
- Find all users who have overdue tasks

### Step 1: The User Class

```python
from datetime import date 

class User: 
    def __init__(self, name, email):
        self.name = name
        self.email = email
        self.tasks = []

    def add_task(self, task):
        self.tasks.append(task)

    def __repr__(self):
        return f"User({self.name}, {len(self.tasks)} tasks)"
```

## Step 2: The dictionary

Now instead of a plain list of users, we store them in a dictionary keyed by emails:

**Why?**:

Email as the key: Because loopup becomes O(1), regardless

```python
user_db = {}

alice = User("alice", "alice@company.com")
bob = User("bob", "bob@company.com")

user_db["alice@company.com"] = alice 
user_db["bob@company.com"] = bob
```

#### Task 1: Add this task data and write a function that returns all users with at least one overdue task:

```python
alice.add_task({"title": "Fix login bug", "due": date(2025, 1, 1), "done": False})
alice.add_task({"title": "Write docs",    "due": date(2026, 12, 1), "done": False})
bob.add_task({"title": "Deploy service", "due": date(2025, 3, 1), "done": True})
```

#### task 2: get_overdue_users function.

- Loop through all users in user_db
- For each user, check if they have at least one task where done is False and the due date is before today
- Return a list of those users

```python
overdue_list = []

for user in user_db.values():
    for task in user.tasks:
        if task["done"]== False and task["due"] < date.today():
            overdue_list.append(user)
            break       # so that user doesnt appear twice if he has 2 due tasks

print(overdue_list)
```

#### Task 3: Error Handling

Right now if you do this:

```python
user = user_db["nobody@company.com"]
```

Your entire program crashes with a KeyError. In a real backend that means the API returns a 500 error to the user — which is bad.

Your job is to write a `function get_user(email)` that:

- Returns the user if they exist

- Returns a meaningful error message if they don't — not a crash

```python
def get_user(email):
    
    user = user_db.get(email)

    if user is None:
        return {"error": "User not found"}, 404
    
    return user, 200
    
        
print(get_user("alice@company.com"))
print(get_user("nobody@company.com"))
```
